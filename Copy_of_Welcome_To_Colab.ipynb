```
"""
STATISTICAL FOUNDATIONS OF MODEL TRAINING
Titanic Survival Prediction - All Tier 1 Concepts

 HOW TO USE THIS NOTEBOOK:
1. Open in Google Colab
2. Click Runtime > Run All (or run each cell one by one)
3. Save the visualizations for your blog
4. Read the explanations to understand each concept

No coding knowledge required - just click and learn!
"""

# ============================================================================
# SETUP: Install and Import Libraries
# ============================================================================
print(" Setting up libraries...")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.utils import resample
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print(" Setup complete!\n")


# ============================================================================
# SECTION 1: LOAD & EXPLORE TITANIC DATA
# ============================================================================
print("="*70)
print(" SECTION 1: LOADING TITANIC DATASET")
print("="*70)

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(f"\n Dataset Shape: {df.shape[0]} passengers, {df.shape[1]} features")
print(f"\n Target Variable: Survived (0 = Died, 1 = Survived)")
print(f"   - Survivors: {df['Survived'].sum()} ({df['Survived'].mean()*100:.1f}%)")
print(f"   - Deaths: {(1-df['Survived']).sum()} ({(1-df['Survived'].mean())*100:.1f}%)")

# Prepare data
print("\n Preparing data...")
# Select relevant features and handle missing values
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']
df_clean = df[features + ['Survived']].copy()
df_clean['Age'].fillna(df_clean['Age'].median(), inplace=True)
df_clean['Fare'].fillna(df_clean['Fare'].median(), inplace=True)
df_clean['Sex'] = df_clean['Sex'].map({'male': 0, 'female': 1})
df_clean.dropna(inplace=True)

X = df_clean[features]
y = df_clean['Survived']

print(f" Clean dataset: {len(X)} passengers ready for analysis\n")

# Show sample data
print(" Sample Data:")
print(df_clean.head(10))


# ============================================================================
# SECTION 2: BIAS-VARIANCE TRADEOFF
# ============================================================================
print("\n" + "="*70)
print(" SECTION 2: BIAS-VARIANCE TRADEOFF")
print("="*70)
print("""
 CONCEPT EXPLANATION:
- BIAS: Error from overly simple models (underfitting)
- VARIANCE: Error from overly complex models (overfitting)
- GOAL: Find the sweet spot between bias and variance

We'll train Decision Trees with different depths:
- Depth 1-2: High bias (too simple)
- Depth 3-5: Balanced (just right)
- Depth 10+: High variance (too complex, memorizes training data)
""")

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train models with different complexity
depths = [1, 2, 3, 5, 7, 10, 15, 20]
train_scores = []
test_scores = []

print("\n Training models with different complexities...\n")

for depth in depths:
    model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model.fit(X_train, y_train)

    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)

    train_scores.append(train_score)
    test_scores.append(test_score)

    print(f"Depth {depth:2d}: Train Acc = {train_score:.3f}, Test Acc = {test_score:.3f}")

# Visualize Bias-Variance Tradeoff
plt.figure(figsize=(12, 6))
plt.plot(depths, train_scores, 'o-', label='Training Accuracy', linewidth=2, markersize=8)
plt.plot(depths, test_scores, 's-', label='Test Accuracy', linewidth=2, markersize=8)
plt.xlabel('Model Complexity (Tree Depth)', fontsize=12, fontweight='bold')
plt.ylabel('Accuracy', fontsize=12, fontweight='bold')
plt.title('Bias-Variance Tradeoff: How Model Complexity Affects Performance',
          fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

# Add annotations
best_idx = test_scores.index(max(test_scores))
plt.annotate('Sweet Spot!\nBest Test Performance',
             xy=(depths[best_idx], test_scores[best_idx]),
             xytext=(depths[best_idx]+3, test_scores[best_idx]-0.05),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=10, color='red', fontweight='bold')

plt.annotate('High Bias\n(Underfitting)', xy=(2, 0.7), fontsize=10, color='blue', fontweight='bold')
plt.annotate('High Variance\n(Overfitting)', xy=(17, 0.65), fontsize=10, color='orange', fontweight='bold')

plt.tight_layout()
plt.savefig('bias_variance_tradeoff.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n Key Insight: Best test accuracy at depth {depths[best_idx]} with {test_scores[best_idx]:.3f}")
print("   - Shallow trees: High bias (miss patterns)")
print("   - Deep trees: High variance (memorize noise)")
print("   - Medium depth: Balanced performance")


# ============================================================================
# SECTION 3: CROSS-VALIDATION
# ============================================================================
print("\n" + "="*70)
print(" SECTION 3: CROSS-VALIDATION")
print("="*70)
print("""
 CONCEPT EXPLANATION:
Single train/test split can be misleading - you might get lucky or unlucky!

CROSS-VALIDATION (K-Fold):
1. Split data into K parts (folds)
2. Train K times, each time using different fold as test
3. Average the results for reliable estimate
4. See how stable your model performance is
""")

# Use optimal depth from previous section
optimal_depth = depths[best_idx]

# Single split score
single_model = DecisionTreeClassifier(max_depth=optimal_depth, random_state=42)
single_model.fit(X_train, y_train)
single_score = single_model.score(X_test, y_test)

# Cross-validation scores
cv_scores = cross_val_score(single_model, X, y, cv=5, scoring='accuracy')

print(f"\n Comparing Single Split vs Cross-Validation (Depth={optimal_depth}):\n")
print(f"Single Train/Test Split:  {single_score:.4f}")
print(f"\n5-Fold Cross-Validation:")
for i, score in enumerate(cv_scores, 1):
    print(f"  Fold {i}: {score:.4f}")
print(f"\n  Mean:    {cv_scores.mean():.4f}")
print(f"  Std Dev: {cv_scores.std():.4f}")
print(f"  Range:   {cv_scores.min():.4f} to {cv_scores.max():.4f}")

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart of CV scores
ax1.bar(range(1, 6), cv_scores, alpha=0.7, color='steelblue', edgecolor='black')
ax1.axhline(cv_scores.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {cv_scores.mean():.3f}')
ax1.axhline(single_score, color='green', linestyle='--', linewidth=2, label=f'Single Split: {single_score:.3f}')
ax1.set_xlabel('Fold Number', fontsize=12, fontweight='bold')
ax1.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax1.set_title('Cross-Validation: Performance Across Different Folds', fontsize=13, fontweight='bold')
ax1.legend()
ax1.set_ylim([0.7, 0.9])
ax1.grid(axis='y', alpha=0.3)

# Box plot comparison
ax2.boxplot([cv_scores], labels=['5-Fold CV'], patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red', linewidth=2))
ax2.scatter([1], [single_score], color='green', s=200, zorder=3, label='Single Split', marker='D')
ax2.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax2.set_title('Reliability: CV Shows Performance Variability', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('cross_validation.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n Key Insight: CV gives more reliable estimate than single split")
print(f"   - Variability: {cv_scores.std():.4f} (shows uncertainty)")
print(f"   - Single split might be lucky/unlucky - CV averages this out")


# ============================================================================
# SECTION 4: CONFIDENCE INTERVALS (BOOTSTRAPPING)
# ============================================================================
print("\n" + "="*70)
print(" SECTION 4: CONFIDENCE INTERVALS")
print("="*70)
print("""
 CONCEPT EXPLANATION:
"My model is 82% accurate" - but how confident are you?

BOOTSTRAPPING:
1. Resample your data many times (with replacement)
2. Train model on each sample
3. Get distribution of scores
4. Calculate 95% confidence interval

This shows: "I'm 95% confident accuracy is between X% and Y%"
""")

# Bootstrap confidence intervals
n_bootstrap = 100
bootstrap_scores = []

print(f"\n Running {n_bootstrap} bootstrap iterations...")

for i in range(n_bootstrap):
    # Resample data
    X_boot, y_boot = resample(X, y, random_state=i)
    X_train_boot, X_test_boot, y_train_boot, y_test_boot = train_test_split(
        X_boot, y_boot, test_size=0.3, random_state=i
    )

    # Train and evaluate
    model = DecisionTreeClassifier(max_depth=optimal_depth, random_state=i)
    model.fit(X_train_boot, y_train_boot)
    score = model.score(X_test_boot, y_test_boot)
    bootstrap_scores.append(score)

bootstrap_scores = np.array(bootstrap_scores)

# Calculate confidence interval
ci_lower = np.percentile(bootstrap_scores, 2.5)
ci_upper = np.percentile(bootstrap_scores, 97.5)
mean_score = bootstrap_scores.mean()

print(f"\n Bootstrap Results ({n_bootstrap} iterations):")
print(f"   Mean Accuracy:  {mean_score:.4f}")
print(f"   95% CI:         [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"   Interpretation: We're 95% confident the true accuracy is between {ci_lower:.1%} and {ci_upper:.1%}")

# Visualize
plt.figure(figsize=(12, 6))
plt.hist(bootstrap_scores, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(mean_score, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_score:.3f}')
plt.axvline(ci_lower, color='green', linestyle='--', linewidth=2, label=f'95% CI Lower: {ci_lower:.3f}')
plt.axvline(ci_upper, color='green', linestyle='--', linewidth=2, label=f'95% CI Upper: {ci_upper:.3f}')
plt.xlabel('Accuracy', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold')
plt.title('Bootstrap Distribution: Quantifying Model Performance Uncertainty', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)

# Shade CI region
plt.axvspan(ci_lower, ci_upper, alpha=0.2, color='green', label='95% Confidence Interval')

plt.tight_layout()
plt.savefig('confidence_intervals.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n Key Insight: Confidence intervals show uncertainty in your estimate")
print(f"   - Narrow CI = More confident in results")
print(f"   - Wide CI = More uncertainty, need more data")


# ============================================================================
# SECTION 5: HYPOTHESIS TESTING
# ============================================================================
print("\n" + "="*70)
print(" SECTION 5: HYPOTHESIS TESTING")
print("="*70)
print("""
 CONCEPT EXPLANATION:
"Model B is better than Model A" - but is it REALLY better, or just lucky?

HYPOTHESIS TESTING:
- Null Hypothesis (H0): Models are equally good
- Alternative (H1): Model B is actually better
- P-value: Probability that difference is just random luck
- If p < 0.05: Reject H0, difference is "statistically significant"

We'll compare two models and calculate if the difference is real.
""")

# Compare two models: shallow vs optimal depth
model_a = DecisionTreeClassifier(max_depth=3, random_state=42)
model_b = DecisionTreeClassifier(max_depth=optimal_depth, random_state=42)

# Get cross-validation scores for both
print("\n Comparing Model A (depth=3) vs Model B (depth={})...".format(optimal_depth))

scores_a = cross_val_score(model_a, X, y, cv=10, scoring='accuracy')
scores_b = cross_val_score(model_b, X, y, cv=10, scoring='accuracy')

# Perform paired t-test
t_stat, p_value = stats.ttest_rel(scores_b, scores_a)

print(f"\n Comparison Results:")
print(f"\nModel A (Depth=3):")
print(f"   Mean Accuracy: {scores_a.mean():.4f}  {scores_a.std():.4f}")
print(f"\nModel B (Depth={optimal_depth}):")
print(f"   Mean Accuracy: {scores_b.mean():.4f}  {scores_b.std():.4f}")
print(f"\nStatistical Test:")
print(f"   Difference:  {(scores_b.mean() - scores_a.mean()):.4f}")
print(f"   T-statistic: {t_stat:.4f}")
print(f"   P-value:     {p_value:.4f}")

if p_value < 0.05:
    print(f"\n CONCLUSION: Model B is STATISTICALLY SIGNIFICANTLY better (p < 0.05)")
    print(f"    The improvement is real, not just random luck!")
else:
    print(f"\n  CONCLUSION: No significant difference (p >= 0.05)")
    print(f"    The difference might just be random variation")

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Box plot comparison
box_data = [scores_a, scores_b]
bp = ax1.boxplot(box_data, labels=['Model A\n(Depth=3)', f'Model B\n(Depth={optimal_depth})'],
                 patch_artist=True)
bp['boxes'][0].set_facecolor('lightcoral')
bp['boxes'][1].set_facecolor('lightgreen')
ax1.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax1.set_title('Model Comparison: Distribution of CV Scores', fontsize=13, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)

# Add significance indicator
if p_value < 0.05:
    y_max = max(scores_a.max(), scores_b.max()) + 0.02
    ax1.plot([1, 2], [y_max, y_max], 'k-', linewidth=2)
    ax1.text(1.5, y_max + 0.005, f'p = {p_value:.4f} *', ha='center', fontweight='bold')

# Paired differences
differences = scores_b - scores_a
ax2.hist(differences, bins=10, alpha=0.7, color='purple', edgecolor='black')
ax2.axvline(0, color='red', linestyle='--', linewidth=2, label='No Difference')
ax2.axvline(differences.mean(), color='green', linestyle='--', linewidth=2,
           label=f'Mean Diff: {differences.mean():.4f}')
ax2.set_xlabel('Accuracy Difference (B - A)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Frequency', fontsize=12, fontweight='bold')
ax2.set_title('Paired Differences: Is Model B Consistently Better?', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('hypothesis_testing.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n Key Insight: P-values tell you if improvements are real or luck")
print(f"   - p < 0.05: Significant (less than 5% chance it's random)")
print(f"   - p >= 0.05: Not significant (could be random variation)")


# ============================================================================
# SUMMARY
# ============================================================================
print("\n" + "="*70)
print(" PROJECT COMPLETE - SUMMARY")
print("="*70)
print("""
You've now demonstrated all 4 Tier 1 Statistical Concepts:

1.  BIAS-VARIANCE TRADEOFF
   - Simple models: High bias (underfit)
   - Complex models: High variance (overfit)
   - Sweet spot: Balanced complexity

2.  CROSS-VALIDATION
   - Single split can mislead
   - K-fold CV gives reliable estimate
   - Shows performance variability

3.  CONFIDENCE INTERVALS
   - Quantifies uncertainty in results
   - "I'm 95% confident accuracy is between X and Y"
   - Bootstrap method for estimation

4.  HYPOTHESIS TESTING
   - Tests if improvements are real
   - P-value < 0.05 = statistically significant
   - Prevents false conclusions

 All visualizations saved! Use them in your blog post.

 For your blog, explain:
- What each concept means (in simple terms)
- Why it matters for real ML projects
- What you discovered in this project
- When to use each technique

Good luck with your Medium post! 
""")

print("\n Files saved:")
print("   - bias_variance_tradeoff.png")
print("   - cross_validation.png")
print("   - confidence_intervals.png")
print("   - hypothesis_testing.png")
```
